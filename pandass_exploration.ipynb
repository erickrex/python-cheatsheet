```python
capitals = pd.DataFrame(
    [
    ['Ngerumuld', 391, 1.87],
    ['Vatican City', 826, 100],
    ['Yaren', 1100, 10.91],
    ['Funafuti', 4492, 45.48],
    ['City of San Marino', 4493]    
    ],
    index = ['Palau', 'Vatican', 'Nauru', 'Tuvalu', 'San Marino'],
    columns = ['Capital', 'Population', 'Percentage']
)
```

```python
capitals
```

```python
capitals.loc['Nauru', 'Population'] #row based indexing
```

```python
capitals['Population']['Nauru']  #chain indexing drills down starting with the column, IT TAKES 2 OPERATIONS
```

```python
capitals.loc['Palau':'Nauru', ['Population', 'Percentage']] #rows and then columns
```

```python
capitals.loc[['Palau', 'Nauru'], :'Percentage'] #rows and then columns
```

```python
capitals.loc[['San Marino', 'Vatican']]
```

```python
capitals.iloc[[4,1,2]] #rows by position
```

```python
capitals.iloc[[4,1], 1:] #rows by position, strictly location based so even with an integer as the name of the index iloc will still refer to the position and not the label index
```

```python
capitals.iloc[:,2] 
```

```python
capitals[[True, True, True, True, False]]
```

```python
capitals['Percentage'] > 25 # returns a list/series so it can be used in an indexing operation
```

```python
capitals[capitals['Percentage']>25] 
```

```python
grades = pd.DataFrame([[6,4], [7,8], [6,7], [6,5], [5,2]],
                        index = ['Mary', 'John', 'Ann', 'Pete', 'Laura'],
                        columns= ['test_1', 'test_2'])
```

```python
grades
```

```python
grades['test_2'] <= grades['test_1']
```

```python
grades[grades['test_2'] <= grades['test_1']]
```

```python
grades.mean() > 5.5 #series of booleans
```

```python
grades.loc[:, grades.mean() > 5.5]
```

```python
grades
```

```python
 grades.loc[['Laura', 'John'], 'test_2'] +=1
```
